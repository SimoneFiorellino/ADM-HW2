```python
import pandas as pd
dataset=pd.read_csv("2019-Oct.csv", header="infer", parse_dates=["event_time"],date_parser=pd.to_datetime)
```

```python
dataset.head(20)
```

```python
dataset_purchase=dataset[dataset["event_type"]=="purchase"]
```

```python
dataset_purchase.head(10)
```

```python
dataset_view=dataset[dataset["event_type"]=="view"]
```

```python
dataset_view.head(10)
```

```python
dataset_viewed_products=dataset_view.groupby(["product_id","category_code"]).product_id.count()\
.sort_values(ascending=False)
dataset_most_viewed=dataset_viewed_products[0:99]
dataset_most_viewed
```

```python
dataset_category_views=dataset_most_viewed.groupby("category_code").sum().sort_values(ascending=False)
dataset_category_views
```

```python
import matplotlib.pyplot as plt
dataset_category_views.plot.barh(figsize=(16,9),\
                                 title="categories views of the most 100 trending products in october",\
                                  color="lightgreen", edgecolor="yellow", alpha=0.5)
plt.show()
```

```python
dataset_sold_products=dataset_purchase.groupby(["product_id","category_code"]).product_id.count()\
.sort_values(ascending=False)

dataset_most_sold_products=dataset_sold_products[0:99]
```

```python
dataset_category_sales=dataset_most_sold_products.groupby("category_code").sum().sort_values(ascending=False)
dataset_category_sales
```

```python
import matplotlib.pyplot as plt

dataset_category_sales.plot.barh(figsize=(16,9),\
                                 title="categories sales of the most 100 trending products in october",\
                                  color="lightblue", edgecolor="yellow", alpha=0.5)
plt.show()
```

```python
series_category_code=dataset_category_sales.index.to_series()
```

```python
i=0
while i<len(dataset_category_sales):
    category=series_category_code[i]
    dataset_new=dataset[dataset["category_code"]==category]
    result=dataset_new.groupby(["product_id","category_code"]).product_id.count()\
    .sort_values(ascending=False).head(10)
    print(category.upper())
    print(result)
    i=i+1
```
