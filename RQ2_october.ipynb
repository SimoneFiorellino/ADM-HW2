In the following cells datasets of october and november are imported. The original dataset has 9 columns. In the imported dataset has been added an additional column, called main category, in which each category.subcategory (e.g. computers(category).notebooks(subcategory)) is splited so that the main category can be used for further analysis.

**october**

```python
import pandas as pd
dataset_oct=pd.read_csv("2019-Oct.csv", header="infer", parse_dates=["event_time"],date_parser=pd.to_datetime,nrows=100000)
dataset_oct=dataset_oct.dropna(subset=["category_code"])
sub_categories=dataset_oct["category_code"].to_list()
print(dataset_oct.shape)
categories=[]
for i in range(len(sub_categories)):
    sub_category=sub_categories[i]
    string=""
    for char in sub_category:
        if char!=".":
            string=string+char
        else:
            break
    categories.append(string)
print(len(categories))
dataset_oct["main category"]=categories

dataset_oct
```

```python
import pandas as pd
dataset_nov=pd.read_csv("/Users/eugeniobaldo/Desktop/2019-Nov.csv", header="infer", parse_dates=["event_time"],date_parser=pd.to_datetime, nrows=100000)
dataset_nov=dataset_nov.dropna(subset=["category_code"])
sub_categories1=dataset_nov["category_code"].to_list()
print(dataset_nov.shape)
categories1=[]
for i in range(len(sub_categories1)):
    sub_category1=sub_categories1[i]
    string1=""
    for char1 in sub_category1:
        if char1!=".":
            string1=string1+char1
        else:
            break
    categories1.append(string1)
print(len(categories1))
dataset_nov["main category"]=categories1

dataset_nov
```

**What are the categories of the most trending products overall? For each month visualize this information through a plot showing the number of sold products per category.**

**october**

```python
dataset_purchase_oct=dataset_oct[dataset_oct["event_type"]=="purchase"]
```

```python
dataset_purchase_oct.head(10)
```

```python
dataset_sold_products_oct=dataset_purchase_oct.groupby(["product_id","main category"]).product_id.count()\
.sort_values(ascending=False)

dataset_most_sold_products_oct=dataset_sold_products_oct[0:99]

dataset_most_sold_products_oct
dataset_category_sales_oct=dataset_most_sold_products_oct.groupby("main category").sum().sort_values(ascending=False)
print(dataset_category_sales_oct)
```

```python
import matplotlib.pyplot as plt
plt.figure()
dataset_category_sales_oct.plot.barh(figsize=(16,9),\
                                 title="categories sales of the most 100 trending products in october",\
                                  color="lightblue", edgecolor="yellow", alpha=0.5)

plt.show()
```
